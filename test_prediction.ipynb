Loading the Modules Needed

```python
from ultralytics import YOLO
import cv2
from matplotlib import pyplot as plt
```

Loading the Test Image

```python
img_name = "./test.png"
img = cv2.imread(f"{img_name}")
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

Making the Prediction

```python
model = YOLO("./model/face_mask_model.pt")
pred = model.predict(img)[0]
```

```python
img_wth_box = pred.plot()
```

Model Result

```python
plt.imshow(cv2.cvtColor(img_wth_box, cv2.COLOR_BGR2RGB))
plt.show()
```
