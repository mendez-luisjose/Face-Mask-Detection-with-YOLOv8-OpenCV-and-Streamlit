Loading the Modules Needed

```python
from ultralytics import YOLO
import cv2
```

Loading the Model

```python
MODEL_DIR = "./model/face_mask_model.pt"

model = YOLO(MODEL_DIR)
```

```python
wCam, hCam = 640, 480
```

Start Making the Predictions in Real Time

```python
cap = cv2.VideoCapture(0)
cap.set(3, wCam)
cap.set(4, hCam)

while True :
    success, frame = cap.read()

    results = model.predict(frame)
    annotations = results[0].plot()

    cv2.imshow("Image", annotations)

    cv2.waitKey(1)

    if cv2.waitKey(10) & 0xFF == ord('q'):
        cap.release()
        cv2.destroyAllWindows()
        break
```
